```python
import pandas as pd 
```

```python
df = pd.read_csv('df.csv', delimiter=';')
df.head()
```

```python
# Remove pontos e notação científica convertendo para inteiros
df['CNS'] = df['CNS'].astype(float).apply(lambda x: '{:.0f}'.format(x))
df['Cross'] = df['Cross'].astype(float).apply(lambda x: '{:.0f}'.format(x))
df['telefone'] = df['telefone'].apply(lambda x: f'+{x}' if not str(x).startswith('+') else str(x))
df.columns = df.columns.str.strip()
```

```python
df.dropna(inplace=True)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import urllib.parse
import pandas as pd

# Abre o navegador e acessa o WhatsApp Web
navegador = webdriver.Chrome()
navegador.get('https://web.whatsapp.com/')

# Espera o carregamento do WhatsApp
while len(navegador.find_elements(By.ID, 'side')) < 1:
    time.sleep(1)

# Limpa colunas e índice
df.columns = df.columns.str.strip()
df.reset_index(drop=True, inplace=True)

# Corrige números de telefone (evita erro float)
df['telefone'] = df['telefone'].astype(str).str.replace('.0', '', regex=False)

# Envia mensagens
for i, mensagem in enumerate(df['Mensagem']):
    nome = df.loc[i, 'Nome']
    telefone = df.loc[i, 'telefone']
    texto = urllib.parse.quote(f'Oi, {nome}! {mensagem}')
    link = f'https://web.whatsapp.com/send?phone={telefone}&text={texto}'
    navegador.get(link)

    # Espera a conversa carregar
    while len(navegador.find_elements(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span/div/div[2]/div[1]/div[2]/div[1]/p')) < 1:
        time.sleep(1)

    time.sleep(2)

    # Envia a mensagem
    campo = navegador.find_element(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span/div/div[2]/div[1]/div[2]/div[1]/p')
    campo.send_keys(Keys.ENTER)

    time.sleep(10)
```
